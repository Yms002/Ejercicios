Descripció de les columnes:
1. Year 2008
2. Month 1-12
3. DayofMonth 1-31
4. DayOfWeek 1 (Monday) - 7 (Sunday)
5. DepTime actual departure time (local, hhmm)
6. CRSDepTime scheduled departure time (local, hhmm)
7. ArrTime actual arrival time (local, hhmm)
8. CRSArrTime scheduled arrival time (local, hhmm)
9. UniqueCarrier unique carrier code
10. FlightNum flight number
11. TailNum plane tail number: aircraft registration, unique aircraft identifier
12. ActualElapsedTime in minutes
13. CRSElapsedTime in minutes
14. AirTime in minutes
15. ArrDelay arrival delay, in minutes: A flight is counted as "on time" if it operated less than 15 minutes later the scheduled time shown in the carriers' Computerized Reservations Systems (CRS).
16. DepDelay departure delay, in minutes
17. Origin origin IATA airport code
18. Dest destination IATA airport code
19. Distance in miles
20. TaxiIn taxi in time, in minutes
21. TaxiOut taxi out time in minutes
22. Cancelled *was the flight cancelled
23. CancellationCode reason for cancellation (A = carrier, B = weather, C = NAS, D = security)
24. Diverted 1 = yes, 0 = no
25. CarrierDelay in minutes: Carrier delay is within the control of the air carrier. Examples of occurrences that may determine carrier delay are: aircraft cleaning, aircraft damage, awaiting the arrival of connecting passengers or crew, baggage, bird strike, cargo loading, catering, computer, outage-carrier equipment, crew legality (pilot or attendant rest), damage by hazardous goods, engineering inspection, fueling, handling disabled passengers, late crew, lavatory servicing, maintenance, oversales, potable water servicing, removal of unruly passenger, slow boarding or seating, stowing carry-on baggage, weight and balance delays.
26. WeatherDelay in minutes: Weather delay is caused by extreme or hazardous weather conditions that are forecasted or manifest themselves on point of departure, enroute, or on point of arrival.
27. NASDelay in minutes: Delay that is within the control of the National Airspace System (NAS) may include: non-extreme weather conditions, airport operations, heavy traffic volume, air traffic control, etc.
28. SecurityDelay in minutes: Security delay is caused by evacuation of a terminal or concourse, re-boarding of aircraft because of security breach, inoperative screening equipment and/or long lines in excess of 29 minutes at screening areas.
29. LateAircraftDelay in minutes: Arrival delay at an airport due to the late arrival of the same aircraft at a previous airport. The ripple effect of an earlier delay at downstream airports is referred to as delay propagation.

```python
# Mirem quiens aerolinies tenen més retard:
retardaero = delay_df.groupby('uniquecarrier')['meandelay'].sum()
print(retardaero)
fig111 = retardaero.plot(kind='bar', title='Retard acumulat per aerolinia')
plt.ylabel('Retard acumulat en minuts')
fig111.figure.savefig(r'Plots\06_Retard_Acumulat_per_aerolinia.png')
```

```python
#El vol més llarg i el vol amb més retras són:
maxdistance = delay_df.loc[delay_df['distance'].idxmax()]
print("El vol més llarg ha estat desde {} fins a {}".format(maxdistance['origin'], maxdistance['dest'])) # EWR correspon a Newark i HNL correspon a Honolulu!
maxdelay = delay_df.loc[delay_df['meandelay'].idxmax()]
print("El vol amb més delay ha estat el {} del {} del {}, amb un delay de {} minuts, {} hores!!!".format(maxdelay['dayofmonth'], maxdelay['month'], maxdelay['year'], maxdelay['meandelay'], (maxdelay['meandelay']/60)))
# 41 hores de delay?? no té pinta de ser correcte!
```

```python
# Afegim una columna que ens indiqui si el vol ha tingut delay (True) o no (False) a l'arribada. 
# Es considera que el vol s'ha retrassat si a l'arribada (arrdelay) hi ha +15 minuts respecte el previst.
delay_df['delayed'] = (delay_df['arrdelay'] > 15)
delay_df['delayed'].describe()
```

```python
delay_df[['distance','airtime']].describe().round(3)
```

```python
# Creem una nova columna amb la velocitat mitjana de cada vol (velocitat = distancia / temps)
# la columna 'airtime' està en minuts, dividim entre 60 per tenir en h
# i la columna 'distance' està en milles, multipliquem per 1,60934 per passar-ho a km
delay_df['velocitat'] = ((delay_df['distance'] * 1.60934) / (delay_df['airtime'] / 60)).round(3)
delay_df['velocitat'].describe().round(3)
```

Resumeix gràficament el data set DelayedFlights.csv

Crea almenys una visualització per:

```python
# Crearem una visualització diferent de la variable categòrca UniqueCarrier: amb una 'pie chart', veurem els vols realitzats per cada aerolinia
plt.rcParams.update({'font.size': 12, 'figure.figsize': (8, 6)}) #establim la mida dels plots

explode = (vols == vols.max()) * 0.1 #fem servir una matriu de booleans per donarli valor al valor màxim i poder "separarlo" de la resta

fig1, ax1 = plt.subplots()
wedges, texts, autotexts = ax1.pie(vols, explode = explode, labels = vols.index, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1.legend(wedges, vols.index,
          title="Aerolinies",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=6, weight="bold")

ax1.set_title("Vols per aerolinia")
plt.show()
fig1.savefig(r'Plots\07_Vols_per_aerolinia_PiePlot.png')

```

```python
# Representarem ara la variable numèrica 'arrdelay'. Farem servir un histograma, en que el retard s'agruparà cada 10 minuts.

fig2 = sns.histplot(delay_df, x = 'arrdelay', binwidth=10, kde = True, color = 'k')
#sns.displot(delay_df, x = 'arrdelay', bins=50, color='k')
plt.title('Arrival delay in minutes')
plt.ylabel('Frequency')
plt.xlabel('Arrival Delay (in minutes)')
plt.show()
fig2.figure.savefig(r"Plots\08_Arrival_Delay_HistPlot.png")
```

Podem veure que la gran majoria de Delay està entre alguns minuts abans de l'hora prevista, i 4 hores (240 minuts) aprox. Es tracta d'una distribució sesgada a la dreta (Right Skewed): hi ha pocs delays massa grans, la majoria son delays relativament curts. Cal tenir en compte també els possibles valors atípics (més de 2400 minuts de retard són més de 40 hores de retard!!). Ho analitzarem amb un Box Plot:

```python
# Creem un Box Plot
sns.set(style="darkgrid") #darkgrid, whitegrid, dark, white, and ticks
fig3 = sns.boxplot(x=delay_df['arrdelay'], color='palegreen', fliersize=0.5, linewidth=0.5, width=0.5)
plt.title('BoxPlot Arrival Delay')
plt.xlabel('Arrival Delay (in minutes)')
plt.show()
fig3.figure.savefig(r"Plots\09_Arrival_Delay_BoxPlot.png")
```

Podem apreciar amb més detall el que deiem abans: el IQR (interquartile range) està entre -60 i +180 aprox (entre una hora abans de l'hora prevista d'arribada i 3 hores de retard)

```python
# Creem la visualització d'una variable numèrica i una categòrica (ArrDelay i UniqueCarrier). Farem servir Violin Plot i Box Plot:
fig4 = plt.figure(figsize=(12, 12))
ax1 = fig4.add_subplot(2, 1, 1)
sns.violinplot(x=delay_df['uniquecarrier'], y=delay_df['arrdelay'], scale='count')
#scale = 'count' fixa l'ample de cada violí segons el nombre d'observacions
plt.title('Arrival delay by Airline in minutes - Violin Plot')
plt.ylabel('Arrival Delay (in minutes)')
plt.xlabel('Airline')
ax2 = fig4.add_subplot(2, 1, 2)
sns.boxplot(x=delay_df['uniquecarrier'], y=delay_df['arrdelay'], fliersize=0.5)
plt.title('Arrival delay by Airline in minutes - Box Plot')
plt.ylabel('Arrival Delay (in minutes)')
plt.xlabel('Airline')
plt.show()
fig4.savefig(r"Plots\10_Arrival_Delay_by_Airline_ViolinPlot_and_BoxPlot.png")
```

Amb la comparativa entre Violin Plot i  Box Plot veiem que WN, l'aerolinia que més vols fa, és de les que menys delay té. Per contra, l'aerolinia B6 que és de les que menys vols realitza, es de les que té el IQR més gran, per tant de les que acumula més retard.

```python
fig5 = sns.catplot(data=delay_df, kind='bar', x='uniquecarrier', y='arrdelay')
plt.title('Mean Arrival delay by Airline in minutes - Bar Plot')
plt.ylabel('Mean Arrival Delay (in minutes)')
plt.xlabel('Airline')
plt.show()
fig5.savefig(r'Plots\11_Mean_Arrival_by_Airline_BarPlot.png')
#delay_df.groupby(['uniquecarrier']).mean()['arrdelay'] 
```

Amb el Barplot veiem la mitjana de 'arrdelay' de cada companyia i podem observar que, les aerolinies B6 y YV que al boxplot eren de les que tenien el IQR més gran, son les que més minuts de delay de mitja tenen.

```python
#Dues variables numèriques (ArrDelay i DepDelay). 
#Per treballar amb dues o més variables numèriques farem servir un scatterplot, combinat amb histogrames.
#sns.scatterplot(x = delay_df['arrdelay'], y = delay_df['depdelay'], hue=delay_df['uniquecarrier']) #scatterplot només
fig6 = sns.JointGrid(data = delay_df, x = 'arrdelay', y = 'depdelay', hue='uniquecarrier', height=8, ratio=4, space=.05)
fig6.plot(sns.scatterplot, sns.histplot) # scatterplot i histograma combinats
plt.show()
fig6.savefig(r'Plots\12_ScatterPlot_and_Histogram.png') #guardem la gràfica
```
