Continua amb el conjunt de dades de tema esportiu i genera una mostra estratificada i una mostra utilitzant SMOTE (Synthetic Minority Oversampling Technique).

```python
normalizacion_antes = df.team_abbreviation.value_counts(normalize=True).sort_index()
# Vemos que puede ser interesante generar una muestra estratificada por equipos
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test = train_test_split(df, test_size=0.1, stratify=df[['team_abbreviation']])
print(f'X_test shape: {X_test.shape}, X_train shape: {X_train.shape}\n')
normalizacion_despues = X_test.team_abbreviation.value_counts(normalize=True).sort_index()
```

Vemos que la forma de la muestra estratificada es del 10% de la población total. A continuación comprobaremos si todos los equipos aparecen correctamente representados:

```python
comparativa_normalizacion = pd.concat([normalizacion_antes,normalizacion_despues], axis = 1)
comparativa_normalizacion.columns = [['Antes de estratificar','Después de estratificar']]
comparativa_normalizacion['Diferencia'] = normalizacion_antes - normalizacion_despues
comparativa_normalizacion
```

Podemos observar que la muestra estratificada conserva correctamente los porcentajes de cada equipo en la población inicial: las variaciones son menores a un 0,00%

A continuación, extraeremos una muestra utilizando el método SMOTE:

```python
# Hemos instalado el paquete imbalanced-learn con: conda install -c conda-forge imbalanced-learn
# import imblearn #lo importamos
from imblearn.over_sampling import SMOTE
```

```python
# Para poder utilizar este método, nos centraremos en si los jugadores son estadounidenses (0) o extranjeros (1)
# muy probablemente la gran mayoria de jugadores sean americanos, por lo que los extranjeros contarán con una representación menor,
# por lo que según que tipo de muestreo que se realice, podría tener una baja representación
usa_counts_before = df.loc[df.country == 'USA', 'country'].count()
df2 = df.copy()
print(f'USA counts antes: {usa_counts_before}')
df2['country'] = df['country'].apply(lambda x: 0 if x == 'USA' else 1)

usa_counts_after = df2.loc[df2.country == 0, 'country'].count()
print(f'USA counts después: {usa_counts_after}')
```

```python
# Contamos cuantos 0 (USA) y 1 (otros) hay:
df2.country.value_counts()
```

```python
# Eliminamos las columnas con str:
columns = ['player_name', 'team_abbreviation', 'college', 'draft_year', 'draft_round', 'draft_number', 'season']
df2 = df2.drop(columns=columns)
```

```python
# Con SMOTE, queremos igualar esos 1 a la cantidad de 0 que hay
oversample = SMOTE()
X, y = oversample.fit_resample(df2, df2[['country']])
X.country.value_counts()
```

Continua amb el conjunt de dades de tema esportiu i genera una mostra utilitzant el mètode Reservoir sampling.

Reservoir sampling: se trata de un algoritmo para escoger aleatoriamente k muestras de una lista de n ítems, donde n suele ser un número muy grande o desconocido. El algoritmo solo pasa una vez por los n ítems y no "mira atrás" en la lista. 

```python
import random # importamos random para generar un número aleatorio
```

```python
# en nuestro caso, n = len(df) y k escogeremos 100
n = len(df)
k = 100
print(f'El tamaño de la poblaclión es {n} y el tamaño de la muestra será {k}')
```

```python
# Creamos una lista vacia en la que se irán introduciendo los ítems de la muestra
# código adaptado del de la siguiente url: https://towardsdatascience.com/coding-interview-questions-reservoir-sampling-687ff5b6822b
reservorio = []
i = 0
for item in df.index:               # iteramos en los índices del df (población)
    i = i + 1 
    if len(reservorio) < k:         # si la lista reservorio tiene menos valores que los que queremos de muestra
        reservorio.append(item)     # añadimos el item a la lista reservorio
    else:                           
        j = int(random.random() * n) # generamos números aleatorios 
        if j < k:                   # si el número aleatorio es menor que k, 
            reservorio[j] = item    # substituimos el item en la posición aleatoria j
reservorio
```

```python
type(reservorio)
```

```python
# Creamos la muestra con los indices obtenidos mediante reservoir sampling
df_reservoir = df.iloc[reservorio]
df_reservoir
```
