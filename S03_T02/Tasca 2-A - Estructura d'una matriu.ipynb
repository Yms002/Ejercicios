### - Exercici 1
Crea un np.array d'una dimensió, que inclogui l'almenys 8 nombres sencers, data type int64. Mostra la dimensió i la forma de la matriu. .

```python
# Primer, hem d'importar numpy:
import numpy as np
# utilitzem el codi de la tasca 2 per generar una llista aleatoria de 10 nombres:
import random as r
llista_aleatoria = []
for i in range(0,10):
    n = r.randint(1,100)
    llista_aleatoria.append(n)
print("La llista aleatoria obtinguda és: " + str(llista_aleatoria))

# Fem servir np.array per transsformar aquesta llista aleatoria en una matriu:
array_aleatoria = np.array(llista_aleatoria)
print("L'array_aleatoria és: " + str(array_aleatoria))
print("El type de la matriu és: " + str(type(array_aleatoria)) + " i el type dels valors de la matriu és: " + str(type(array_aleatoria[0])))
print("La dimensió de la matriu és: " + str(array_aleatoria.ndim))
print("La forma de la matriu és: " + str(array_aleatoria.shape))
```

### - Exercici 2
De la matriu de l'exercici 1, calcula el valor mitjà dels valors introduïts i resta la mitjana resultant de cada un dels valors de la matriu.

```python
# farem servir np.mean per la mitjana i np.median pel valor mig
mitjana = round(np.mean(array_aleatoria),2)
mig = np.median(array_aleatoria)
print("La mitjana és: " + str(mitjana) + "; i el valor mig és: " + str(mig))

# Restem la mitjana als valors de la matriu:
resta = array_aleatoria - mitjana
print("Si restem la mitjana (" + str(mitjana) + ") a cada un dels valors de la llista, obtenim la nova matriu: " + str(resta))
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as  plt

foto_Baloo = mpimg.imread('Baloo and Qu.jpeg')
plt.imshow(foto_Baloo)
plt.axis('off') # eliminamos los ejes que pinta matplotlib
plt.show("La foto en imatge")
print("La foto com a array: " + str(foto_Baloo))
print("Dimensions de l'array: " + str(foto_Baloo.ndim) + "; Shape de l'array: " + str(foto_Baloo.shape))

'''
#Com veiem, la matriu té les dimensions de 1080 x 1080 (mida de la imatge en pixels) x 3 (canals RGB)
plt.imshow(foto_Baloo[:,:,0])
plt.title("Canal 'Red'")
plt.figure()
plt.imshow(foto_Baloo[:,:,1])
plt.title("Canal 'Green'")
plt.figure()
plt.imshow(foto_Baloo[:,:,2])
plt.title("Canal 'Blue'")
'''

#Creem còpies per no perdre l'original
baloo_red_green = np.copy(foto_Baloo)
baloo_red_blue = np.copy(foto_Baloo)
baloo_green_blue = np.copy(foto_Baloo)

#eliminem els canals que volem treure igualant-lo a 0:

baloo_red_green[:,:,2] = 0
plt.title("Sense canal blau")
plt.imshow(baloo_red_green)
plt.figure()
baloo_red_blue[:,:,1] = 0
plt.title("Sense canal verd")
plt.imshow(baloo_red_blue)
plt.figure()
baloo_green_blue[:,:,0] = 0
plt.title("Sense canal vermell")
plt.imshow(baloo_green_blue)

#guardem les imatges:
mpimg.imsave("Baloo_red_green.png", baloo_red_green)
mpimg.imsave("Baloo_red_blue.png", baloo_red_blue)
mpimg.imsave("Baloo_green_blue.png", baloo_green_blue)
```
